```python
# Correlation among features (useful to mention in findings)
corr = df[['X','Y','Z']].corr().values
labels = ['X','Y','Z']

plt.figure()
im = plt.imshow(corr, interpolation='nearest')
plt.xticks(range(len(labels)), labels)
plt.yticks(range(len(labels)), labels)
plt.title('Correlation Matrix (X, Y, Z)')
plt.colorbar(im, fraction=0.046, pad=0.04)
for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, f"{corr[i, j]:.2f}", ha='center', va='center')
plt.tight_layout()
plt.show()
```

```python
# Boxplots: distribution of X, Y, Z across steps (13 boxes per feature)
import matplotlib.pyplot as plt

for feature in ['X','Y','Z']:
    plt.figure()
    # Create a list of arrays, one per class, to boxplot
    data_by_step = [df.loc[df['Step']==s, feature].values for s in sorted(df['Step'].unique())]
    plt.boxplot(data_by_step, showfliers=False)  # no custom colors, one figure per plot
    plt.title(f'Boxplot of {feature} by Step')
    plt.xlabel('Step')
    plt.ylabel(feature)
    plt.xticks(ticks=range(1, len(data_by_step)+1), labels=sorted(df['Step'].unique()))
    plt.tight_layout()
    plt.show()
```

```python
# Pairwise scatter plots to inspect class separability in 2D
# Matplotlib will auto-assign colors; we don't set any
pairs = [('X','Y'), ('X','Z'), ('Y','Z')]
steps_sorted = sorted(df['Step'].unique())

for a, b in pairs:
    plt.figure()
    for s in steps_sorted:
        subset = df[df['Step']==s]
        plt.scatter(subset[a].values, subset[b].values, s=12, alpha=0.7, label=f"Step {s}")
    plt.xlabel(a); plt.ylabel(b)
    plt.title(f'{a} vs {b} by Step')
    # Too many legend entries can clutter; toggle next line if you want it:
    plt.legend(title='Step', fontsize=8, ncol=3, loc='best')
    plt.tight_layout()
    plt.show()
```

```python
# 3D view to see overall structure (optional but insightful)
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for s in steps_sorted:
    subset = df[df['Step']==s]
    ax.scatter(subset['X'].values, subset['Y'].values, subset['Z'].values, s=12, alpha=0.8, label=f"Step {s}")
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
ax.set_title('3D Scatter of (X, Y, Z) by Step')
# Legends in 3D can overlap; comment out if it blocks points:
ax.legend(title='Step', fontsize=8, ncol=2, loc='best')
plt.tight_layout()
plt.show()
```

```python
# Signal-to-noise ratio per feature: variance of class means / mean of within-class variances
import numpy as np

snr = {}
for feature in ['X','Y','Z']:
    means_by_class = df.groupby('Step')[feature].mean().values
    within_vars = df.groupby('Step')[feature].var().values
    between_var = np.var(means_by_class, ddof=1)
    within_var_mean = np.nanmean(within_vars)
    snr[feature] = between_var / within_var_mean if within_var_mean > 0 else np.inf

print("Class separability (higher is better):")
for k,v in snr.items():
    print(f"  {k}: SNR = {v:.3f}")
```

```python
# Class centroids = mean coordinates per step (great to describe clusters)
centroids = df.groupby('Step')[['X','Y','Z']].mean().rename(columns={
    'X':'X_centroid','Y':'Y_centroid','Z':'Z_centroid'
})
display(centroids)
```

# STEP 4: Step 4: Classification Model Development/Engineering - 20 Marks


Prepare the data to create three classification models (based on ML algorithms). The dataset needs to be split into training and testing categories to develop the models. For each ML model, utilize grid search cross-validation to assess the hyperparameters that give you the best results. You are required to explain your selected choice of classification algorithms. In addition to the three classification models with grid search cross-validation, you must make one model based on using RandomizedSearchCV. this will provide another method of determining the best hyperparameters to optimize your results.
Package(s) required: scikit-learn - All ML models can be accessed from the sklearn package


```python
# CELL A  imports, reproducibility, split, helpers

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# --- Reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# --- Feature/target selection
FEATURES = ['X','Y','Z']
TARGET = 'Step'
X = df[FEATURES].copy()
y = df[TARGET].astype(int).copy()

# --- Stratified split to preserve class ratios
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
)

# --- Utility: fit grid/randomized search and return best estimator + best CV score
def fit_search(name, estimator, param_grid_or_dist, search='grid', cv=5):
    if search == 'grid':
        searcher = GridSearchCV(estimator, param_grid_or_dist, cv=cv, scoring='accuracy', n_jobs=-1, verbose=0)
    else:
        searcher = RandomizedSearchCV(estimator, param_grid_or_dist, n_iter=40, cv=cv, scoring='accuracy',
                                      n_jobs=-1, random_state=RANDOM_STATE, verbose=0)
    searcher.fit(X_train, y_train)
    return name, searcher.best_estimator_, searcher.best_score_, searcher.best_params_

# --- Utility: evaluate on held-out test set
def evaluate(name, model):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"\n{name}  Test Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred, digits=4))
    plt.figure()
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f"Confusion Matrix  {name}")
    plt.tight_layout(); plt.show()
    return acc
```

## KNN with scaling inside the pipeline


```python
# CELL B  KNN with scaling inside the pipeline

pipe_knn = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', KNeighborsClassifier())
])

grid_knn = {
    'clf__n_neighbors': [3, 5, 7, 9],
    'clf__weights': ['uniform', 'distance'],
    'clf__p': [1, 2]  # Manhattan vs Euclidean
}

name_knn, best_knn, cv_knn, params_knn = fit_search("KNN (Grid)", pipe_knn, grid_knn, search='grid')
print("KNN best params:", params_knn)
print("KNN best CV accuracy:", cv_knn)
acc_knn = evaluate(name_knn, best_knn)
```

## SVM RBF with scaling inside the pipeline

```python
pipe_svm = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', SVC(kernel='rbf', decision_function_shape='ovr', random_state=RANDOM_STATE))
])

grid_svm = {
    'clf__C': [0.3, 1, 3, 10, 30],
    'clf__gamma': ['scale', 0.3, 0.1, 0.03, 0.01]
}

name_svm, best_svm, cv_svm, params_svm = fit_search("SVM RBF (Grid)", pipe_svm, grid_svm, search='grid')
print("SVM best params:", params_svm)
print("SVM best CV accuracy:", cv_svm)
acc_svm = evaluate(name_svm, best_svm)
```

## Random Forest (no scaling)

```python
# Faster RF grid that still finds good params

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    random_state=RANDOM_STATE,
    n_jobs=-1,
    bootstrap=True,
    max_samples=0.8  # subsample for speed; remove if you prefer full bootstrap
)

grid_rf = {
    'n_estimators': [100, 200, 300],   # trimmed from 600
    'max_depth': [None, 10, 20],       # trimmed
    'min_samples_split': [2, 5],       # trimmed
    'min_samples_leaf': [1, 2],        # trimmed
    'max_features': ['sqrt']           # standard & fastest here (3 features anyway)
}

name_rf, best_rf, cv_rf, params_rf = fit_search("Random Forest (Grid)", rf, grid_rf, search='grid')
print("RF best params:", params_rf)
print("RF best CV accuracy:", cv_rf)
acc_rf = evaluate(name_rf, best_rf)
```

## Gradient Boosting with RandomizedSearchCV

```python
## FAST Gradient Boosting with RandomizedSearchCV (early stopping + trimmed search)
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

gb = GradientBoostingClassifier(
    random_state=RANDOM_STATE,
    subsample=0.8,            # stochastic GB speeds up & regularizes
    n_iter_no_change=10,      # early stopping
    validation_fraction=0.1,  # 10% of training set held out for early stopping
    tol=1e-3
)

param_dist_gb = {
    'n_estimators': np.arange(50, 201, 25),     # 50..200 (no need for 600 here)
    'learning_rate': [0.03, 0.05, 0.1],         # sensible small set
    'max_depth': [2, 3],                        # shallow trees work best for GB
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2, 4],
    # 'subsample': [0.8]  # fixed above (dont search it)
}

rs_gb = RandomizedSearchCV(
    gb,
    param_distributions=param_dist_gb,
    n_iter=20,                 # was 40  halve the work
    cv=5,
    scoring='accuracy',
    n_jobs=-1,                 # parallelize across CV folds / candidates
    random_state=RANDOM_STATE,
    verbose=1
)
rs_gb.fit(X_train, y_train)

name_gb, best_gb, cv_gb, params_gb = "Gradient Boosting (Randomized)", rs_gb.best_estimator_, rs_gb.best_score_, rs_gb.best_params_
print("GB best params:", params_gb)
print("GB best CV accuracy:", cv_gb)
acc_gb = evaluate(name_gb, best_gb)
```

# Step 5: Model Performance Analysis - 20 Marks
Compare the overall performance of each model based on f1 score, precision and accuracy. You are required to provide an explanation to what these metrics mean and which metric to prioritize for this use-case. Based on the selected model, create a confusion matrix to visualize the performance of your model. Include the confusion matrix in the report as well. Package(s) required: scikit-learn (Metrics, Confusion Matrix)


```python
from sklearn.metrics import f1_score, precision_score, accuracy_score

# Store predictions
y_pred_knn = best_knn.predict(X_test)
y_pred_svm = best_svm.predict(X_test)
y_pred_rf  = best_rf.predict(X_test)
y_pred_gb  = best_gb.predict(X_test)

def scores(y_true, y_pred, name):
    return {
        "Model": name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average='weighted'),
        "F1_score": f1_score(y_true, y_pred, average='weighted')
    }

results_metrics = pd.DataFrame([
    scores(y_test, y_pred_knn, "KNN"),
    scores(y_test, y_pred_svm, "SVM RBF"),
    scores(y_test, y_pred_rf,  "Random Forest"),
    scores(y_test, y_pred_gb,  "Gradient Boosting")
])

display(results_metrics.sort_values("Accuracy", ascending=False))
```

```python
# STEP 6  Stacked Model Performance Analysis

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, f1_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd

# 1) Build the stack: SVM (RBF) + Random Forest -> meta-learner Logistic Regression
stack = StackingClassifier(
    estimators=[
        ("svm", best_svm),   # pipeline with scaler + SVC(RBF)
        ("rf",  best_rf)     # tuned RandomForest
    ],
    final_estimator=LogisticRegression(max_iter=1000, multi_class="auto"),
    passthrough=True,        # include original features along with base learners' outputs
    cv=5,                    # out-of-fold predictions for meta-learner
    n_jobs=-1
)

# 2) Fit the stacked model
stack.fit(X_train, y_train)

# 3) Evaluate
y_pred_stack = stack.predict(X_test)

acc_stack = accuracy_score(y_test, y_pred_stack)
prec_stack = precision_score(y_test, y_pred_stack, average="weighted")
f1_stack = f1_score(y_test, y_pred_stack, average="weighted")

print(f"Stacked (SVM+RF)  Test Accuracy: {acc_stack:.4f}")
print(f"Stacked (SVM+RF)  Weighted Precision: {prec_stack:.4f}")
print(f"Stacked (SVM+RF)  Weighted F1: {f1_stack:.4f}")

# 4) Confusion matrix for the stacked model (add this image to your report)
plt.figure()
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_stack, cmap="Blues", colorbar=True)
plt.title("Confusion Matrix  Stacked (SVM+RF  Logistic Regression)")
plt.tight_layout()
plt.show()

# 5) Compare against individual models from Step 5 (assuming y_pred_* already computed)
comparison = pd.DataFrame([
    {"Model": "KNN",                "Accuracy": accuracy_score(y_test, best_knn.predict(X_test)),
                                    "Precision": precision_score(y_test, best_knn.predict(X_test), average="weighted"),
                                    "F1": f1_score(y_test, best_knn.predict(X_test), average="weighted")},
    {"Model": "SVM RBF",            "Accuracy": accuracy_score(y_test, best_svm.predict(X_test)),
                                    "Precision": precision_score(y_test, best_svm.predict(X_test), average="weighted"),
                                    "F1": f1_score(y_test, best_svm.predict(X_test), average="weighted")},
    {"Model": "Random Forest",      "Accuracy": accuracy_score(y_test, best_rf.predict(X_test)),
                                    "Precision": precision_score(y_test, best_rf.predict(X_test), average="weighted"),
                                    "F1": f1_score(y_test, best_rf.predict(X_test), average="weighted")},
    {"Model": "Gradient Boosting",  "Accuracy": accuracy_score(y_test, best_gb.predict(X_test)),
                                    "Precision": precision_score(y_test, best_gb.predict(X_test), average="weighted"),
                                    "F1": f1_score(y_test, best_gb.predict(X_test), average="weighted")},
    {"Model": "Stacked (SVM+RF)",   "Accuracy": acc_stack, "Precision": prec_stack, "F1": f1_stack}
]).sort_values("Accuracy", ascending=False).reset_index(drop=True)

display(comparison)
```
