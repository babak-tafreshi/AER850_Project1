#Step 1: Data Processing - 2 Marks
Read data from a csv file and convert that into a dataframe, which will allow for all further analysis and data manipulation.
Package(s) required: Pandas


```python
# Import pandas
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("Project1_Data.csv")

# Show the first 5 rows to check the data
print(df.head(),"\n", )

# Check general info (data types, missing values)
print("\n", df.info())

# Basic statistics (mean, min, max, etc.)
print("\n", df.describe())
```

```python
# Statistical analysis per class (Step)
class_stats = df.groupby('Step')[['X','Y','Z']].agg(
    count=('X','count'),
    X_mean=('X','mean'), X_std=('X','std'), X_min=('X','min'), X_median=('X','median'), X_max=('X','max'),
    Y_mean=('Y','mean'), Y_std=('Y','std'), Y_min=('Y','min'), Y_median=('Y','median'), Y_max=('Y','max'),
    Z_mean=('Z','mean'), Z_std=('Z','std'), Z_min=('Z','min'), Z_median=('Z','median'), Z_max=('Z','max'),
)
display(class_stats)
```

```python
# Overall numeric summary to reference in text
overall_stats = df[['X','Y','Z']].describe().T
display(overall_stats)
```

```python
# Class centroids = mean coordinates per step (to describe clusters)
centroids = df.groupby('Step')[['X','Y','Z']].mean().rename(columns={
    'X':'X_centroid','Y':'Y_centroid','Z':'Z_centroid'
})
display(centroids)
```
